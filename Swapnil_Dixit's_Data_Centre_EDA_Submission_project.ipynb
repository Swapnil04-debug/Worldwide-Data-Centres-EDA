The rapid expansion of cloud computing, artificial intelligence, and digital services has significantly increased the demand for data centers worldwide. The exploratory analysis focuses on identifying patterns and relationships between operational and environmental metrics. The project also investigates the relationship between estimated capacity and electricity consumption, PUE and energy efficiency, water usage and surrounding water stress levels, as well as trends across different years.

The primary objective of this project is not only to summarize the dataset but also to generate actionable insights into the operational efficiency and sustainability of modern data centers. The findings can help identify high-capacity facilities, evaluate the effectiveness of different cooling technologies, understand regional resource consumption patterns, and highlight opportunities for improving energy and water efficiency. Furthermore, the project demonstrates the practical application of Python libraries such as Pandas, NumPy, Matplotlib, and Seaborn for real-world data cleaning, transformation, visualization, and exploratory analysis.

The insights generated from this analysis can support informed decision-making, encourage sustainable infrastructure planning, and serve as a strong foundation for future predictive modeling or optimization studies related to energy-efficient data center management.

````python
# 1. Capacity (MW)
df['Capacity_Category'] = pd.cut( df['Estimated_Capacity_MW'],bins=[0, 5, 20, 50, np.inf],labels=['Small', 'Medium', 'Large', 'Hyperscale']
)

# 2. PUE (industry standard)
df['PUE_Category'] = pd.cut(df['PUE'],bins=[0, 1.3, 1.6, 2.0, np.inf], labels=['Excellent', 'Good', 'Average', 'Poor'])

# 3. WUE (quantile based)
df['WUE_Category'] = pd.qcut(df['WUE_L_per_kWh'],q=3,labels=['Low', 'Medium', 'High']
)

# 4. Electricity Usage (quantile based)
df['Electricity_Usage_Category'] = pd.qcut(df['Daily_Electricity_Usage_MWh'],q=3,labels=['Low', 'Medium', 'High'])

# 5. Water Usage (quantile based)
df['Water_Usage_Category'] = pd.qcut(df['Daily_Water_Usage_Gallons'],q=3, labels=['Low', 'Medium', 'High'])
````
